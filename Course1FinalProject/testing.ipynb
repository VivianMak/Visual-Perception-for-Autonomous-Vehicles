```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_yaw_angle(path):
    # Calculate differences in x and y coordinates
    dx = np.diff(path[:, 0])  # Differences in x coordinates
    dy = np.diff(path[:, 1])  # Differences in y coordinates

    # Calculate yaw angle using arctan2
    yaw_angles = np.arctan2(dy, dx)

    return yaw_angles

# Example path (x, y coordinates)
path = np.array([[0, 0],
                 [1, 1],
                 [2, 2],
                 [3, 1],
                 [4, 0]])

# Calculate yaw angle
yaw_angles = calculate_yaw_angle(path)

# Plot the path
plt.figure(figsize=(8, 6))
plt.plot(path[:, 0], path[:, 1], marker='o', label='Path')

# Plot arrows representing yaw angles
for i in range(len(path) - 1):
    plt.arrow(path[i, 0], path[i, 1], np.cos(yaw_angles[i]), np.sin(yaw_angles[i]), 
              head_width=0.1, head_length=0.1, fc='red', ec='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Yaw Angle of Path')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
